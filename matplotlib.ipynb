```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


months = ['Jan','Feb','March','APRIL','MAY','JUNE']
energy_consumption = [1200,1300,1100,1600,1400,1700]

#create a line plot
plt.plot(months,energy_consumption,marker='o', color='y', linestyle='--')

#add ttitle and labels
plt.title('Energy Consumption')
plt.xlabel('Months')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
#simple data for energy consumption by different renewable sources
energy_sources = ['Solar','wind','hydropower','biomass']
energy_values = [1200,1300,1500,1300]

#create a bar chart
plt.bar(energy_sources,energy_values,color='green')

plt.title('Energy Consumption by renewable energy source')
plt.xlabel('Energy Sources')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
#pie chart to show energy consumption distribution
plt.pie(energy_values,labels=energy_sources,autopct='%1.1f%%')

plt.title('Energy consumption share by source' )
plt.show()
```

```python
#simple data for energy consumption and carbon emission
carbon_emissions = [200,500,450,300]

plt.scatter(energy_values, carbon_emissions, color='b')

plt.title('Energy Consumption vs Carbon Emission')
plt.xlabel('Energy Consumption(Mwh)')
plt.ylabel('Carbon Emission(kg CO2)')
plt.show()
```

```python
plt.bar(energy_sources,energy_values,color='green', edgecolor='black')


plt.title('Customized Energy Consumption by source',fontsize=14 ,fontweight='bold')
plt.xlabel('Energy Sources', fontsize=12)
plt.ylabel('Energy Consumption', fontsize=12)


plt.grid(True,linestyles='--')
plt.legend(['energy consumption'], loc='upper right')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Solar": [1500, 3000, 1200],
    "Wind": [1200, 1300, 1500],
    "Hydropower": [1000, 1200, 900],
    "Biomass": [800, 1000, 700]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()

#plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')


#add title
plt.title('Correlation between energy sources')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Boxplot of the dataset')
plt.show()
```

```python
sns.violinplot(data=df)
plt.title('Violinplot of energy consumption distribution')
plt.show()
```
