```python
#PROBLEM4
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "Date": pd.date_range(start="2021-10-01",periods=10),

    "Measurement":[16,13,14,12,np.nan,np.nan,np.nan,8,7,5]
})
print(df)
```
